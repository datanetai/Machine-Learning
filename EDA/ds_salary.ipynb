```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('data/ds_salaries.csv')
data.drop(['Unnamed: 0'], axis=1, inplace=True)
data.head()
```

```python
categorial_columns = data.select_dtypes(include=['object']).columns
numerical_columns = data.select_dtypes(include=['float64','int64']).columns
print("Categorial columns: ", categorial_columns)
print("Numerical columns: ", numerical_columns)
```

```python
# distribution for numerical columns
plt.figure(figsize=(10,10))
for i,col in enumerate(numerical_columns):
    plt.subplot(2,2,i+1)
    sns.histplot(data[col])
    plt.title(col)
plt.show()
```

```python
# distribution for categorial columns
categorial_columns = ['experience_level', 'employment_type','company_size']
plt.figure(figsize=(10,10))
for i,col in enumerate(categorial_columns):
    plt.subplot(2,2,i+1)
    sns.countplot(x=col, data=data)
    plt.title(col)
```

```python
# unique values for job_title
len(data['job_title'].unique())
```

```python
# top 5 job_title
colors = ['#0072B2', '#009E73', '#D55E00', '#CC79A7', '#F0E442']
data['job_title'].value_counts().head(5).plot(kind='barh', color=colors)
plt.title('Top 5 job_title')
plt.show()
```

```python
# top 10 employee_residence
colors = ['#0072B2', '#009E73', '#D55E00', '#CC79A7', '#F0E442', '#56B4E9', '#E69F00', '#F0E442', '#009E73', '#D55E00']
data['employee_residence'].value_counts().head(10).plot(kind='bar', color=colors)
plt.title('Top 10 employee_residence')
plt.xticks(rotation=45)
plt.show()
```

```python
# salary distribution for different experience levels
sns.boxplot(x='experience_level',y='salary_in_usd',data=data)
plt.title('Salary Distribution for Different Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Salary')
plt.show()
```
